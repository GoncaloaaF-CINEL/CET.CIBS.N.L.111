5 Cria uma função chamada media que receba dois números e devolva a média.

```python
def media(num1, num2):
    soma = num1 + num2
    media = soma / 2
    return media

res = media(4,10)
print(res)
```

6 Cria uma função chamada par_ou_impar que receba um número e devolva:

   * "Par" se for par
   * "Ímpar" se for ímpar

```python
def par_ou_impar(num):
    if num % 2 == 0:
        return "par"
    else:
        return "impar"

par_ou_impar(6)
```


7 Cria uma função chamada maior que receba dois números e devolva o maior.


```python
def maior(n1, n2):
    if n1 > n2:
        return n1
    else:
        return n2

print(maior(4,6))
print(maior(31,3))
```

8 Cria uma função chamada positivo_negativo que receba um número e devolva:
   * "Positivo"
   * "Negativo"
   * "Zero"


```python
def positivo_negativo(n1):
    if n1 > 0:
        return "Positivo"
    elif n1 < 0:
        return "Negativo"
    else:
        return "Zero"
```


9 Cria uma função chamada tabuada que receba um número e imprima a tabuada desse número (de 1 a 10).

```python
def tabuada(n):
    for i in range(1,11):
        calc = n * i
        print(f"{n} x {i} = {calc}")
```

```python
tabuada("dasd")
```

```python
# Exercio 9

# Função para formatar a apresentação (estética)

def formatar_titulo(texto):
    print("-" * 20)
    print(texto.center(20))
    print("-" * 20)

formatar_titulo(f"Tabuada do {3}")

# gerar a tabuada

def gerar_tabuada(n):

    formatar_titulo(f"Tabuada do {n}")

    for i in range(1, 11):

        resultado = n * i
        print(f"{n} x {i:2} = {resultado}")
    print("-" * 20)

#main do Exercio 9

# Pedir o número ao utilizador

while True:
    try:
        num = int(input("Digite um número para ver a tabuada: "))
        gerar_tabuada(num)
        break
    except ValueError:
        print("Erro: Por favor, digite apenas números inteiros.")
```
