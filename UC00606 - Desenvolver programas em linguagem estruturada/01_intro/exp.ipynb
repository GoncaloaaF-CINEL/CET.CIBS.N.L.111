# 4. Faça um Programa que peça as 4 notas bimestrais e mostre a média

```python
nota1 = float(input("nota 1: "))
nota2 = float(input("nota 2: "))
nota3 = float(input("nota 3: "))
nota4 = float(input("nota 4: "))

media = (nota1 + nota2 + nota3 + nota4) / 4
print(f"a media é {media}")
```

```python
nota1 = float(input("nota 1: "))
nota2 = float(input("nota 2: "))
nota3 = float(input("nota 3: "))
nota4 = float(input("nota 4: "))

soma = (nota1 + nota2 + nota3 + nota4)
media = soma / 4
print(f"a media é {media}")
```

```python
nota1 = float(input("nota 1: "))
nota2 = float(input("nota 2: "))
nota3 = float(input("nota 3: "))
nota4 = float(input("nota 4: "))

soma = (nota1 + nota2 + nota3 + nota4)
media = soma / 4
print(f"a media é {media:.2f}")
print(media.__format__(".2f"))
```

5. Faça um Programa que converta metros para centímetros.

```python
valor = float(input("valor em m: "))

cm = valor * 100

print(f"{valor:.2f}m = {cm:.2f}cm ")
```

6. Faça um Programa que peça o raio de um círculo, calcule e mostre sua área

```python
raio_circulo = float(input("raio circulo: "))

area = 3.14 * raio_circulo ** 2

print(f"area = {area:.3f}")
```

```python
import math

raio_circulo = float(input("raio circulo: "))

area = math.pi * raio_circulo ** 2

print(f"area = {area:.3f}")
```

```python
import math

raio_circulo = float(input("raio circulo: "))

area = math.pi * raio_circulo ** 2

print(f"area = {area:.12f}")
```

```python
import math

raio_circulo = float(input("raio circulo: "))

area = math.pi * pow(raio_circulo, 2)

print(f"area = {area:.12f}")
```
