 Faça um Programa que verifique se uma letra digitada é "F" ou "M".

Conforme a letra escrever:

  * F - Feminino,
  * M - Masculino,
  * Sexo Inválido.

```python
gen = input("Digite o seu sexo (M/F): ").upper()

if gen == "M":
    print("Masculino")
elif gen == "F":
    print("Feminino")
else:
    print("Sexo Inválido")
```

Faça um programa para a leitura de duas notas parciais de um aluno.
O programa deve calcular a média alcançada pelo aluno e apresentar:

 * A mensagem "Aprovado", se a média alcançada for maior ou igual a 10;
 * A mensagem "Reprovado", se a média for menor do que 10;
 * A mensagem "Aprovado com Distinção", se a média for igual a 20.

```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))

soma = nota1 + nota2  # <- gastar mais memoria
media = soma / 2

print(f"a media foi {media:.2f}")

if media == 20:
    print("Aprovado com Distinção")
elif media >= 10:
    print("Aprovado")
else:
    print("Reporvado")
```
